```python
import os

print("=== 🔍 EDA : ADFA-LD (SYSCALLS LINUX) ===\n")

# On ajoute le sous-dossier 'ADFA-LD' dans le chemin
TRAIN_DIR = '../data/ADFA-LD/Training_Data_Master/'

try:
    # On liste les fichiers dans le dossier
    fichiers = os.listdir(TRAIN_DIR)
    print(f"📁 {len(fichiers)} fichiers de comportement 'Normal' trouvés.")
    
    if len(fichiers) > 0:
        premier_fichier = os.path.join(TRAIN_DIR, fichiers[0])
        
        # On ouvre et on lit le fichier
        with open(premier_fichier, 'r') as f:
            contenu = f.read().strip()
            
        print(f"\n📄 Aperçu du fichier '{fichiers[0]}' :")
        print(contenu[:100] + " ... (suite coupée)")
        
        # On compte combien d'appels système la machine a fait
        nb_syscalls = len(contenu.split())
        print(f"\n📊 Ce processus a fait {nb_syscalls} appels au noyau Linux.")

except FileNotFoundError:
    print(f"❌ ERREUR : Le dossier {TRAIN_DIR} est introuvable. Vérifie ton arborescence.")
```
